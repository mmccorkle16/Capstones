# Pre-Processing

```python
stemm = PorterStemmer()
lemm = WordNetLemmatizer()
```

```python
def apply_stem(doc):
    doc_split = doc.split(' ')
    stem_doc = ''
    for word in doc_split:
        stem_doc += stemm.stem(word) + ' '
    return stem_doc

def apply_lem(doc):
    doc_split = doc.split(' ')
    lem_v_doc = ''
    for word in doc_split:
        lem_v_doc += lemm.lemmatize(word, pos='v') + ' '
    doc_split = lem_v_doc.split(' ')
    lem_doc = ''
    for word in doc_split:
        lem_doc += lemm.lemmatize(word, pos='a') + ' '
    return lem_doc
```

```python
df['review_stemmed'] = df['Review'].apply(apply_stem)
df['review_lemmed'] = df['Review'].apply(apply_lem)
```

```python
review_index_to_check = 863
print('ORIGINAL REVIEW: ', '\n', df['Review'].iloc[review_index_to_check], '\n')
print('STEMMED REVIEW: ', '\n', df['review_stemmed'].iloc[review_index_to_check], '\n')
print('LEMMED REVIEW: ', '\n', df['review_lemmed'].iloc[review_index_to_check], '\n')
```

```python
#lemmatization is the better approach since it doesn't cut off some words like stemmation did
```

```python
x = df.drop(columns=['Rating'])
y = df['Rating']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train_df = pd.DataFrame(y_train).reset_index(drop=True)
y_test_df = pd.DataFrame(y_test).reset_index(drop=True)
y_train_df.head()
```

```python
tokenizer = RegexpTokenizer(r'[a-zA-Z]+')

tf_idf_vec = TfidfVectorizer(lowercase=True, tokenizer=tokenizer.tokenize, stop_words=stop_words, ngram_range=(1,2), min_df=0.05)

x_lemm_train_tfidf_sparse = tf_idf_vec.fit_transform(x_train['review_lemmed'])
x_lemm_test_tfidf_sparse = tf_idf_vec.transform(x_test['review_lemmed'])

x_lemm_train_tfidf = pd.DataFrame(x_lemm_train_tfidf_sparse.toarray(), columns=tf_idf_vec.get_feature_names())
x_lemm_test_tfidf = pd.DataFrame(x_lemm_test_tfidf_sparse.toarray(), columns=tf_idf_vec.get_feature_names())

x_lemm_train_tfidf.head()
```

```python
x_lemm_train_text = x_lemm_train_tfidf
x_lemm_test_text = x_lemm_test_tfidf
x_lemm_train_text.head()
```
